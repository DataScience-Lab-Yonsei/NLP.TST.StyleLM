# 데이터 확인

```python
movie_df = pd.read_csv("movie_df.csv", encoding="utf-8", index_col=0)
movie_df
```

```python
genre_list = list(set(movie_df['장르']))
genre_list
```

```python
genre_count = dict([(k, 0) for k in genre_list])

for i in range(len(movie_df)):
  genre_count[movie_df.loc[i, "장르"]] += 1

genre_count
```

편의를 위해 크게

코미디 / 드라마 / 액션, 범죄 / (멜로/로맨스)

이렇게 묶인 대본만 사용!

```python
import collections
```

```python
genre_interest = {
    "코미디" : 0,
    "드라마": 1,
    "액션" : 2,
    "범죄" : 2,
    "멜로/로맨스": 3
}

genre_interest
```

```python
import copy
```

```python
def labelling(gen):
  if gen not in list(genre_interest.keys()):
    return -1
  else:
    return genre_interest[gen]
```

```python
movie_label = copy.deepcopy(movie_df)
movie_label['스타일'] = movie_label['장르'].apply(labelling)
movie_label.tail()
```
